```python
import numpy as np
import matplotlib.pyplot as plt

```

```python


d = 100
a = 5

ones = np.ones((d,1))

sigmax = np.eye(d) + a * ones @ ones.T

#b = -1/d
b = 1

#wstar = ( np.eye(d) + b * ones @ ones.T ) / d
wstar = b * ones @ ones.T / d
#wstar = np.random.normal(0,1/np.sqrt(d),(d,d))

w0 = np.random.normal(0,1/np.sqrt(d),(d,d))


alpha = 1/(a*d+1)
ntime = 1000
w = np.zeros((d,d,ntime))
w[:,:,0] = w0
loss = np.zeros((ntime,))
loss[0] = np.trace( ( np.squeeze(w[:,:,0]) - wstar ) @ sigmax @ ( np.squeeze(w[:,:,0]) - wstar ).T ) / 2
for itime in range(ntime-1):
    w[:,:,itime+1] = np.squeeze(w[:,:,itime]) - alpha * sigmax @ ( np.squeeze(w[:,:,itime]) - wstar )
    loss[itime+1] = np.trace( ( np.squeeze(w[:,:,itime+1]) - wstar ) @ sigmax @ ( np.squeeze(w[:,:,itime+1]) - wstar ).T ) / 2


    
    
    
```

```python

plt.plot(loss)
#plt.xlim((0,50))
plt.show()


```
