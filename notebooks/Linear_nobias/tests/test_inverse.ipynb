```python
import numpy as np
import matplotlib.pyplot as plt

import symmetric_methods as sm
```

```python
# MLP structure

L = 2

dim = np.zeros((L+1,))

dim[0] = 9
dim[1] = 7
dim[2] = 5
#dim[3] = 7
#dim[4] = 9
#dim[5] = 8
#dim[6] = 9
#dim[7] = 10
#dim[8] = 11
#dim[9] = 12

dim= dim.astype(int)
```

```python
#Â draw random parameters, test matrix inverse 

phi = np.random.normal(0,1,(L-2,))
PHI1 = np.random.normal(0,1,(dim[0],dim[0]))
PHI1 = PHI1 + PHI1.T
PHIL = np.random.normal(0,1,(dim[L],dim[L]))
PHIL = PHIL + PHIL.T
PSI1 = []
if L==2:
    PSI1 = np.random.normal(0,1,(dim[0],dim[2]))

covg = sm.par2cov(L,dim,phi,PHI1,PSI1,PHIL)
icovg = np.linalg.inv(covg)

phi_inv,PHI1_inv,PSI1_inv,PHIL_inv = sm.cov2par(L,dim,icovg)

phi_invest,PHI1_invest,PSI1_invest,PHIL_invest = sm.inverse(L,dim,phi,PHI1,PSI1,PHIL)

icovgest = sm.par2cov(L,dim,phi_invest,PHI1_invest,PSI1_invest,PHIL_invest)


```

```python

plt.plot(icovg,icovgest,linestyle='none',marker='.')
plt.plot([np.min(icovg), np.max(icovg)],[np.min(icovg), np.max(icovg)])
plt.show()


print(phi_inv)
print('\n')
print(phi_invest)
print('\n')
print(PHI1_inv)
print('\n')
print(PHI1_invest)
print('\n')
print(PSI1_inv)
print('\n')
print(PSI1_invest)
print('\n')
print(PHIL_inv)
print('\n')
print(PHIL_invest)

```
