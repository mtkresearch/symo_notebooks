```python
import numpy as np
import matplotlib.pyplot as plt

import symmetric_methods as sm
```

```python
# MLP structure

L = 2

dim = np.zeros((L+1,))

dim[0] = 9
dim[1] = 7
dim[2] = 5
#dim[3] = 7
#dim[4] = 9
#dim[5] = 8
#dim[6] = 9
#dim[7] = 10
#dim[8] = 11
#dim[9] = 12

dim= dim.astype(int)
```

```python
#Â draw random parameters, test matrix multiplication 

phi_1 = np.random.normal(0,1,(L-2,))
PHI1_1 = np.random.normal(0,1,(dim[0],dim[0]))
PHI1_1 = PHI1_1 + PHI1_1.T
PHIL_1 = np.random.normal(0,1,(dim[L],dim[L]))
PHIL_1 = PHIL_1 + PHIL_1.T
PSI1_1 = []
if L==2:
    PSI1_1 = np.random.normal(0,1,(dim[0],dim[2]))

phi_2 = np.random.normal(0,1,(L-2,))
PHI1_2 = np.random.normal(0,1,(dim[0],dim[0]))
PHI1_2 = PHI1_2 + PHI1_2.T
PHIL_2 = np.random.normal(0,1,(dim[L],dim[L]))
PHIL_2 = PHIL_2 + PHIL_2.T
PSI1_2 = []
if L==2:
    PSI1_2 = np.random.normal(0,1,(dim[0],dim[2]))


covg_1 = sm.par2cov(L,dim,phi_1,PHI1_1,PSI1_1,PHIL_1)
covg_2 = sm.par2cov(L,dim,phi_2,PHI1_2,PSI1_2,PHIL_2)

covg_mult = 0.5 * (covg_1 @ covg_2 + covg_2 @ covg_1)

phi,PHI1,PSI1,PHIL = sm.cov2par(L,dim,covg_mult)

phi_est,PHI1_est,PSI1_est,PHIL_est = sm.matmat(L,dim,phi_1,PHI1_1,PSI1_1,PHIL_1,phi_2,PHI1_2,PSI1_2,PHIL_2)

covg_mult_est = sm.par2cov(L,dim,phi_est,PHI1_est,PSI1_est,PHIL_est)
```

```python

plt.plot(covg_mult,covg_mult_est,linestyle='none',marker='.')
plt.plot([np.min(covg_mult), np.max(covg_mult)],[np.min(covg_mult), np.max(covg_mult)])
plt.show()


print(phi)
print('\n')
print(phi_est)
print('\n')
print(PHI1)
print('\n')
print(PHI1_est)
print('\n')
print(PSI1)
print('\n')
print(PSI1_est)
print('\n')
print(PHIL)
print('\n')
print(PHIL_est)


```
