```python
import numpy as np
import matplotlib.pyplot as plt

import symmetric_methods as sm
```

```python
# MLP structure

L = 7

dim = np.zeros((L+1,))

dim[0] = 3
dim[1] = 4
dim[2] = 5
dim[3] = 6
dim[4] = 7
dim[5] = 8
dim[6] = 9
dim[7] = 10
#dim[8] = 11
#dim[9] = 12

dim= dim.astype(int)
```

```python
#Â draw random parameters, test (symmetric) matrix multiplication 

wd_1 = np.random.normal(0,1,(4,L))
wt_1 = np.random.normal(0,1,(2,L-1))
wo_1 = np.random.normal(0,1,((L-2)*(L-1)//2,))

wd_2 = np.random.normal(0,1,(4,L))
wt_2 = np.random.normal(0,1,(2,L-1))
wo_2 = np.random.normal(0,1,((L-2)*(L-1)//2,))

covg_1 = sm.par2cov(L,dim,wd_1,wt_1,wo_1)
covg_2 = sm.par2cov(L,dim,wd_2,wt_2,wo_2)

covg_mult = 0.5 * (covg_1 @ covg_2 + covg_2 @ covg_1)

wd, wt, wo = sm.cov2par(L,dim,covg_mult)

wdest, wtest, woest = sm.matmat(L,wd_1,wt_1,wo_1,wd_2,wt_2,wo_2)

covg_mult_est = sm.par2cov(L,dim,wdest,wtest,woest)
```

```python

plt.plot(covg_mult,covg_mult_est,linestyle='none',marker='.')
plt.plot([np.min(covg_mult), np.max(covg_mult)],[np.min(covg_mult), np.max(covg_mult)])
plt.show()


for ip in np.arange(0,4):
    print(wd[ip,:])
    print(wdest[ip,:])

for ip in np.arange(0,2):
    print(wt[ip,:])
    print(wtest[ip,:])

print(wo)
print(woest)


```
