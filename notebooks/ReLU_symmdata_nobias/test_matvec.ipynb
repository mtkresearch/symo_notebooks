```python
import numpy as np
import matplotlib.pyplot as plt

import symmetric_methods as sm
```

```python
# MLP structure

L = 7

dim = np.zeros((L+1,))

dim[0] = 3
dim[1] = 4
dim[2] = 5
dim[3] = 6
dim[4] = 7
dim[5] = 8
dim[6] = 9
dim[7] = 10
#dim[8] = 11
#dim[9] = 12

dim= dim.astype(int)
```

```python
#Â draw random parameters, test matrix vector product

wd = np.random.normal(0,1,(4,L))
wt = np.random.normal(0,1,(2,L-1))
wo = np.random.normal(0,1,((L-2)*(L-1)//2,))

covg = sm.par2cov(L,dim,wd,wt,wo)


npar = 0
for iL in range(L):
    npar = npar + dim[iL] * dim[iL+1]
npar = npar.astype(int)

vecinp = np.random.normal(0,1,(npar,))

#vecoutest = sm.matvec(L,dim,wd,wt,wo,vecinp) 
vecoutest = sm.matvec_fast(L,dim,wd,wt,wo,vecinp) 

vecout = covg @ vecinp

```

```python


plt.plot(vecout,vecoutest,linestyle='none',marker='.')
plt.show()



```
