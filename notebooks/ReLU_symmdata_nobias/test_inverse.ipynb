```python
import numpy as np
import matplotlib.pyplot as plt

import symmetric_methods as sm
```

```python
# MLP structure

L = 7

dim = np.zeros((L+1,))

dim[0] = 3
dim[1] = 4
dim[2] = 5
dim[3] = 6
dim[4] = 7
dim[5] = 8
dim[6] = 9
dim[7] = 10
#dim[8] = 11
#dim[9] = 12

dim= dim.astype(int)
```

```python
#Â draw random parameters, test matrix inverse 

wd = np.random.normal(0,1,(4,L))
wt = np.random.normal(0,1,(2,L-1))
wo = np.random.normal(0,1,((L-2)*(L-1)//2,))

covg = sm.par2cov(L,dim,wd,wt,wo)
icovg = np.linalg.inv(covg)

wdinv, wtinv, woinv = sm.cov2par(L,dim,icovg)

wdinvest, wtinvest, woinvest = sm.inverse(L,wd,wt,wo)

icovgest = sm.par2cov(L,dim,wdinvest,wtinvest,woinvest)


```

```python

plt.plot(icovg,icovgest,linestyle='none',marker='.')
plt.plot([np.min(icovg), np.max(icovg)],[np.min(icovg), np.max(icovg)])
plt.show()


for ip in np.arange(0,4):
    print(wdinv[ip,:])
    print(wdinvest[ip,:])

for ip in np.arange(0,2):
    print(wtinv[ip,:])
    print(wtinvest[ip,:])

print(woinv)
print(woinvest)

```
