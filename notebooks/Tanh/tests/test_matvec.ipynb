```python
import numpy as np
import matplotlib.pyplot as plt

import symmetric_methods as sm
```

```python
# MLP structure

L = 2

dim = np.zeros((L+1,))

dim[0] = 5
dim[1] = 5
dim[2] = 5
#dim[3] = 7
#dim[4] = 9
#dim[5] = 8
#dim[6] = 9
#dim[7] = 10
#dim[8] = 11
#dim[9] = 12

dim= dim.astype(int)
```

```python
#Â draw random parameters, test matrix vector product

phi = np.random.normal(0,1,(L-2,))
ttphi = np.random.normal(0,1,(L-1,))
tPhi1 = np.random.normal(0,1,(dim[0],1))
tPhiL = np.random.normal(0,1,(dim[L],1))
PHI1 = np.random.normal(0,1,(dim[0],dim[0]))
PHI1 = PHI1 + PHI1.T
PHIL = np.random.normal(0,1,(dim[L],dim[L]))
PHIL = PHIL + PHIL.T
ttPHIL = np.random.normal(0,1,(dim[L],dim[L]))
ttPHIL = ttPHIL + ttPHIL.T
PSI1 = []
if L==2:
    PSI1 = np.random.normal(0,1,(dim[0],dim[2]))

covg = sm.par2cov(L,dim,phi,ttphi,tPhi1,tPhiL,PHI1,PSI1,PHIL,ttPHIL)


npar = 0
for iL in range(L):
    npar = npar + (dim[iL]+1) * dim[iL+1]
npar = npar.astype(int)

vecinp = np.random.normal(0,1,(npar,))

vecoutest = sm.matvec(L,dim,phi,ttphi,tPhi1,tPhiL,PHI1,PSI1,PHIL,ttPHIL,vecinp) 

vecout = covg @ vecinp

```

```python


plt.plot(vecout,vecoutest,linestyle='none',marker='.')
plt.show()



```
