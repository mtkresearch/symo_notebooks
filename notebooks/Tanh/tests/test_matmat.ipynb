```python
import numpy as np
import matplotlib.pyplot as plt

import symmetric_methods as sm
```

```python
# MLP structure

L = 2

dim = np.zeros((L+1,))

dim[0] = 9
dim[1] = 7
dim[2] = 5
#dim[3] = 7
#dim[4] = 9
#dim[5] = 8
#dim[6] = 9
#dim[7] = 10
#dim[8] = 11
#dim[9] = 12

dim= dim.astype(int)
```

```python
#Â draw random parameters, test matrix multiplication 

phi_1 = np.random.normal(0,1,(L-2,))
ttphi_1 = np.random.normal(0,1,(L-1,))
tPhi1_1 = np.random.normal(0,1,(dim[0],1))
tPhiL_1 = np.random.normal(0,1,(dim[L],1))
PHI1_1 = np.random.normal(0,1,(dim[0],dim[0]))
PHI1_1 = PHI1_1 + PHI1_1.T
PHIL_1 = np.random.normal(0,1,(dim[L],dim[L]))
PHIL_1 = PHIL_1 + PHIL_1.T
ttPHIL_1 = np.random.normal(0,1,(dim[L],dim[L]))
ttPHIL_1 = ttPHIL_1 + ttPHIL_1.T
PSI1_1=[]
if L==2:
    PSI1_1 = np.random.normal(0,1,(dim[0],dim[2]))
    

phi_2 = np.random.normal(0,1,(L-2,))
ttphi_2 = np.random.normal(0,1,(L-1,))
tPhi1_2 = np.random.normal(0,1,(dim[0],1))
tPhiL_2 = np.random.normal(0,1,(dim[L],1))
PHI1_2 = np.random.normal(0,1,(dim[0],dim[0]))
PHI1_2 = PHI1_2 + PHI1_2.T
PHIL_2 = np.random.normal(0,1,(dim[L],dim[L]))
PHIL_2 = PHIL_2 + PHIL_2.T
ttPHIL_2 = np.random.normal(0,1,(dim[L],dim[L]))
ttPHIL_2 = ttPHIL_2 + ttPHIL_2.T
PSI1_2=[]
if L==2:
    PSI1_2 = np.random.normal(0,1,(dim[0],dim[2]))


covg_1 = sm.par2cov(L,dim,phi_1,ttphi_1,tPhi1_1,tPhiL_1,PHI1_1,PSI1_1,PHIL_1,ttPHIL_1)
covg_2 = sm.par2cov(L,dim,phi_2,ttphi_2,tPhi1_2,tPhiL_2,PHI1_2,PSI1_2,PHIL_2,ttPHIL_2)

covg_mult = 0.5 * (covg_1 @ covg_2 + covg_2 @ covg_1)

phi,ttphi,tPhi1,tPhiL,PHI1,PSI1,PHIL,ttPHIL = sm.cov2par(L,dim,covg_mult)

phi_est,ttphi_est,tPhi1_est,tPhiL_est,PHI1_est,PSI1_est,PHIL_est,ttPHIL_est = sm.matmat(L,dim,phi_1,ttphi_1,tPhi1_1,tPhiL_1,PHI1_1,PSI1_1,PHIL_1,ttPHIL_1,phi_2,ttphi_2,tPhi1_2,tPhiL_2,PHI1_2,PSI1_2,PHIL_2,ttPHIL_2)

covg_mult_est = sm.par2cov(L,dim,phi_est,ttphi_est,tPhi1_est,tPhiL_est,PHI1_est,PSI1_est,PHIL_est,ttPHIL_est)
```

```python

plt.plot(covg_mult,covg_mult_est,linestyle='none',marker='.')
plt.plot([np.min(covg_mult), np.max(covg_mult)],[np.min(covg_mult), np.max(covg_mult)])
plt.show()


print(phi)
print('\n')
print(phi_est)
print('\n')
print(ttphi)
print('\n')
print(ttphi_est)
print('\n')
print(tPhi1)
print('\n')
print(tPhi1_est)
print('\n')
print(tPhiL)
print('\n')
print(tPhiL_est)
print('\n')
print(PHI1)
print('\n')
print(PHI1_est)
print('\n')
print(PSI1)
print('\n')
print(PSI1_est)
print('\n')
print(PHIL)
print('\n')
print(PHIL_est)
print('\n')
print(ttPHIL)
print('\n')
print(ttPHIL_est)


```
